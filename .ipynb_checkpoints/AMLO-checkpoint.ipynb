```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
import seaborn as sns
pd.set_option('display.max_rows', None)
```

```python
data = {}

i = 0
for p in tqdm(range(1, 90)):
    url = 'https://lopezobrador.org.mx/page/' + str(p) +'/?s=Versi%C3%B3n+estenogr%C3%A1fica+de+la+conferencia+de+prensa+matutina+del+presidente+Andr%C3%A9s+Manuel+L%C3%B3pez+Obrador'
    d = requests.get(url)
    page = BeautifulSoup(d.text)
    list_of_pages = page.findAll('div', class_='entry-post')

    
    for el in list_of_pages: 
        data[i] = {'date':el.find('span', class_="entry-date").text,
                   'url': el.find('span', class_="entry-date").find('a').get('href'),
                   'title': el.find('h2', class_="entry-title").text}
        i+=1
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
```

```python
df = df.drop(index=df.index[-1])
```

```python
months = {
    'enero': 'January', 'febrero': 'February', 'marzo': 'March', 
    'abril': 'April', 'mayo': 'May', 'junio': 'June', 
    'julio': 'July', 'agosto': 'August', 'septiembre': 'September', 
    'octubre': 'October', 'noviembre': 'November', 'diciembre': 'December'
}

# Function to replace Spanish month names with English month names
def replace_month(date_str):
    for spanish, english in months.items():
        date_str = date_str.replace(spanish, english)
    return date_str
```

```python

# Apply the function to the 'dates' column
df['dates2'] = df['date'].apply(replace_month)

# Convert to datetime
df['dates2'] = pd.to_datetime(df['dates2'], format='%B %d, %Y')

df
```

```python
df.to_csv('conference_links.csv', index=False)
```

```python
df['month'] = df['dates2'].dt.month
df['year'] = df['dates2'].dt.year
```

```python
df_month = df.groupby(['year', 'month'], as_index=False)['dates2'].count().sort_values('dates2')
```

```python
df_month['date'] = pd.to_datetime('15/' + df_month['month'].astype(str).str.zfill(2) +  '/' + df_month['year'].astype(str), dayfirst=True)
```

```python
sns.lineplot(df_month, x='date', y='dates2')
```
