# Creating the dataframe

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
```

```python
#df = df.drop(index=df.index[-1])
```

```python
months = {
    'enero': 'January', 'febrero': 'February', 'marzo': 'March',
    'abril': 'April', 'mayo': 'May', 'junio': 'June',
    'julio': 'July', 'agosto': 'August', 'septiembre': 'September',
    'octubre': 'October', 'noviembre': 'November', 'diciembre': 'December'
}

# Function to replace Spanish month names with English month names
def replace_month(date_str):
    for spanish, english in months.items():
        date_str = date_str.replace(spanish, english)
    return date_str
```

```python
# Apply the function to the 'dates' column
df['dates2'] = df['date'].apply(replace_month)

# Convert to datetime
df['dates2'] = pd.to_datetime(df['dates2'], format='%B %d, %Y')

# Remove time component
df['dates2'] = df['dates2'].dt.normalize()

df['month'] = df['dates2'].dt.month
df['year'] = df['dates2'].dt.year

df_month = df.groupby(['year', 'month'], as_index=False)['dates2'].count().sort_values('dates2')

df_month['date'] = pd.to_datetime('15/' + df_month['month'].astype(str).str.zfill(2) +  '/' + df_month['year'].astype(str), dayfirst=True)
```

```python
sns.lineplot(df_month, x='date', y='dates2')
```
